### Part 1

```python
def search_2ints(input):
    ints = []
    for line in input:
        intsline = []
        for item in line:
            if item.isdigit():
                intsline = intsline + [int(item)]
                #print(intsline)
        ints.append([intsline[0], intsline[-1]])
    return ints
```

```python
tst = search_2ints(inputfile)
print(tst)
```

```python
def list2int(list):
    listofints = []
    for item in list:
        item = int(str(item[0]) + str(item[1]))
        listofints.append(item)
    return listofints

```

```python
listofints = list2int(tst)
print(listofints)
```

```python
summed = np.sum(listofints)
print(summed)
```

### Part 2

```python
digits = ['one','two','three','four','five','six','seven','eight','nine']
```

```python
with open('input.txt', 'r') as f:
    data = f.read().split('\n')
```

```python
print(data)
```

```python
def searchstrandint(data):
    ints = []
    for line in data:
        lineints = []
        for i in range(len(line)):
            if line[i].isdigit():
                lineints.append(int(line[i]))
            else:
                for d in digits:
                    if d in line[i:i+len(d)]:
                        lineints.append(digits.index(d)+1)
        ints.append(lineints)
    return ints

```

```python
test2 = searchstrandint(data)
print(test2)
```

```python
def only2(list):
    newlist = []
    for item in list:
        newlist.append([item[0], item[-1]])
    return newlist
    
```

```python
result = only2(test2)
print(result)
```

```python
result = list2int(result)
print(result)
```

```python
summed2 = np.sum(result)
print(summed2)
```
